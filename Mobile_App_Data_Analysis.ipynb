We have to insert suitable value in NAN of Category

```python
var = mob.iloc[10472][0].upper()
var = var.split()
var
```

```python
cat = []
for i in var:
  for j, k in zip(mob.App_Name, mob.Category):
    if i in j.upper():
      print(f'App : {j} | | Category : {k}')
      cat.append(str(k))
```

```python
random = pd.DataFrame(cat, columns=['Category'])
random
```

```python
random.Category.value_counts()
```

Thus we have found that the Suitable Category for Nan is 'PHOTOGRAPHY'
